```python
edge = pd.read_csv('../data/edge_df_top20000.csv', index_col=0)
```

```python
# 동시출현 값이 20000보다 큰 행만 추출
edge_20000 = edge[edge['val'] > 20000]
```

```python
# edge_20000에 포함되는 단어들만 node 데이터에서 추출
set_nodes = list(set(list(edge['word1']) + list(edge['word2'])))
```

- node의 크기는 언급량
- edge의 두께는 동시출현건수로 설정하여 네트워크 그림

```python
fig, ax = plt.subplots(figsize=(12, 12))
ax.set_facecolor('white')
G = nx.from_pandas_edgelist(edge_20000, 'word1', 'word2', create_using=nx.Graph())
node2 = node.loc[idxs, :]
node2 = node2.sort_values('word').set_index('word')
node2 = node2.reindex(G.nodes())

node2['count'] = pd.Categorical(node2['count'])

node_sizes = [entry*0.027 for entry in node2['count']]
edge_width = list(np.array(edge_20000['val'])*0.0001)
nx.draw_networkx(G, with_labels=True, node_color='#fdb863', 
        node_size=node_sizes, font_family=font_family, font_size = 12,
       font_color = 'black', alpha=0.7, width = edge_width, edge_color = 'gray')
plt.savefig('../results/갓생_networkx_네트워크.svg')
```
