```python
import ast
word_list = list()
```

```python
%%time
for i in file['tokens']:
    i = ast.literal_eval(str(i))
    word_list.append(i)
```

# 추가 불용어(있으면) 제거 및 동시행렬 매트릭스 생성

```python
file.insert(len(file.columns), 'word_list', word_list)
```

```python
def del_stopwords_2(x):
    return [word for word in x if word not in ['하나','여기','이것','동안','얼마','아이','이상']] # not in [리스트] 리스트에 새로운 불용어 추가
```

```python
file['word_list'] = file['word_list'].apply(lambda x : del_stopwords_2(x))
```

```python
file.head()
```

```python
word_list = list(file['word_list'])
```

# 단어 수 세기

```python
%%time
# 워드클라우드용 중복제거 count
word_count = dict()

for words in word_list:
    words = list(set(words))
    for i in words:
        if i not in word_count:
            word_count[i] =1
        else:
            word_count[i] += 1 
```

# 단어 동시 출현빈도
- 한 document에 단어쌍이 한 번이라도 존재하면 +1

```python
from tqdm import tqdm_notebook
from tqdm.notebook import tqdm
```

```python
%%time
count = dict()  #동시출현 빈도가 저장될 dict
# for context in word_list:
for num in tqdm(range(0, len(word_list))):
    context = list(set(word_list[num]))
    for i, a in enumerate(context):
        for b in context[i:]:
            if a == b: continue   #같은 단어의 경우는 세지 않음
            if a > b:    #A, B와 B, A가 다르게 세어지는것을 막기 위해 항상 a < b로 순서 고정
                count[a, b] = count.get((a, b), 0) + 1   #실제로 센다
            else:
                count[b, a] = count.get((b, a), 0) + 1
count
```

```python
count
```

```python
sorted_x = sorted(count.items(), key=lambda kv : kv[1], reverse = True)

top = 2000

unique_terms = []
for k, v in sorted_x[:top]:
    unique_terms+=list(k)
unique_terms = list(set(unique_terms))
unique_terms = sorted(unique_terms)


word_index = dict()
for index, term in enumerate(unique_terms):
    word_index[term] = index

matrix = np.zeros((len(unique_terms), len(unique_terms)))

for k, v in sorted_x[:top]:
    matrix[word_index[k[0]], word_index[k[1]]] = v
    matrix[word_index[k[1]], word_index[k[0]]] = v
    
new_df = pd.DataFrame(matrix, columns=word_index.keys())

reindex = dict()
for i, k in enumerate(word_index.keys()):
    reindex[i] = k

new_df.rename(index = reindex, inplace=True)

new_df.to_csv('./매트릭스_top2000.csv')
```

```python
%%time
edge_df = pd.DataFrame(columns=['word1', 'word2', 'val'])
for i in sorted_x[:20000]:
    edge_df.loc[len(edge_df)] = [i[0][0], i[0][1],i[1]]
```
