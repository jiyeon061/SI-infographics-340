```python
file = pd.read_csv('../data/갓생_tokenized.csv' ,index_col=0)
```

```python
import ast
word_list = list()
```

```python
%%time
for i in file['tokens']:
    i = ast.literal_eval(str(i))
    word_list.append(i)
```

```python
file.insert(len(file.columns), 'word_list', word_list)
```

```python
def del_stopwords_2(x):
    return [word for word in x if word not in ['갓생','하나','여기','이것','동안','얼마','아이','이상']] # not in [리스트] 리스트에 새로운 불용어 추가
```

```python
file['word_list'] = file['word_list'].apply(lambda x : del_stopwords_2(x))
file.head()
```

```python
workout_idx = [idx for idx in file.index if '운동' in str(file.loc[idx, 'content'])]
```

```python
study_idx = [idx for idx in file.index if '공부' in str(file.loc[idx, 'content'])]
```

```python
workout = file.loc[workout_idx, :]
```

```python
study = file.loc[study_idx, :]
```

```python
workout.head()
```

```python
study.head()
```

```python
word_list_workout = list(workout['word_list'])
```

```python
word_list_study = list(study['word_list'])
```

```python
%%time
# 워드클라우드용 중복제거 count
word_count = dict()

for words in word_list_workout:
    words = list(set(words))
    for i in words:
        if i not in word_count:
            word_count[i] =1
        else:
            word_count[i] += 1 
```

```python
sorted_by_value = sorted(word_count.items(), key=lambda x: x[1], reverse=True)
```

```python
# sorted_by_value # 운동
```

```python
df_workout = pd.DataFrame(columns=['word', 'count'])
for i in sorted_by_value:
    df_workout.loc[len(df_workout)] = [i[0], i[1]]
```

```python
# sorted_by_value
```

```python
%%time
# 워드클라우드용 중복제거 count
word_count = dict()

for words in word_list_study:
    words = list(set(words))
    for i in words:
        if i not in word_count:
            word_count[i] =1
        else:
            word_count[i] += 1 
```

```python
sorted_by_value2 = sorted(word_count.items(), key=lambda x: x[1], reverse=True)
```

```python
# sorted_by_value2 # 공부
```

```python
df_study = pd.DataFrame(columns=['word', 'count'])
for i in sorted_by_value2:
    df_study.loc[len(df_study)] = [i[0], i[1]]
```

# 갓생+운동 워드클라우드

```python
font = '../font/KoPub Dotum Bold.ttf'
```

```python
file = pd.read_csv('../data/갓생_운동_wordcount.csv', index_col=0)
df_word = pd.DataFrame(file).head(100)
dic_word=df_word.set_index('word').to_dict()['count']
```

```python
_min = df_word['count'].min()
_max = df_word['count'].max()
# df_word = df_word.head(100)
```

```python
def color_func(word, font_size, position,orientation,random_state=None, **kwargs):
    return("hsl({:d},{:d}%, {:d}%)".format(np.random.randint(13,18),np.random.randint(56,73),np.random.randint(33,65)))
```

```python
def my_tf_color_func(dic_word):
    def my_tf_color_func_inner(word, font_size, position, orientation, random_state=None, **kwargs):
        _max = max(dic_word.values())
        _min = min(dic_word.values())
        return "hsl({:d}, {:d}%, {:d}%)".format(13, 73, int(50*((1-((dic_word[word]-_min)/(_max-_min)))+0.5)))
        # return "hsl({:d}, {:d}%, {:d}%)".format(100, 52, 50*((1-dic_word[word])+0.5))
    return my_tf_color_func_inner
```

```python
from wordcloud import WordCloud
wc = WordCloud(random_state = 1234,         # 난수 고정
               font_path = font,            # 폰트 설정
               width = 5000,                 # 가로 크기
               height = 5000,               # 세로 크기
               background_color= "white")   #배경색  
#                colormap="OrRd")  #글자색
```

```python
%%time
img_wordcloud = wc.generate_from_frequencies(dic_word)
```

```python
%%time   
# 워드 클라우드 만들기
# img_wordcloud = wc.generate_from_frequencies(dic_word)

# 워드 클라우드 출력하기
plt.figure(figsize = (10, 10))  # 가로, 세로 크기 설정
plt.axis('off')                 # 테두리 선 없애기
# plt.imshow(img_wordcloud)       # 워드 클라우드 출력
plt.imshow(img_wordcloud.recolor(color_func = my_tf_color_func(dic_word)),interpolation="bilinear") 
plt.savefig('../results/갓생+운동_워드클라우드.svg', transparent = True)
```
